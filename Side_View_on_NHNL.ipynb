## Study 1:  Spike Bounce Standby Mode

##Is Driving on a Yellow Light Dangerous?##
<img src="SpikeTrade-Traffic-Light.jpg", style="width: 35%; border: 2px solid black;">

### Event Definitions

##Standby##

* %NH - %NL > -20 yesterday
* %NH - %NL <= -20 today

##Bounce##

* %NH - %NL <= -20 yesterday
* %NH - %NL > -20 today

### Historical Results By Holding Horizons

```python
events = OrderedDict()
events['Bounce'] = (d_all.spike_bounce)
events['Standby'] = (d_all.spike_standby)
```

### Event Definitions

##Fresh Bounce##
* Today is a "Spike Bounce" day.
* There was no "Spike Bounce" in the previous 4 days.
